## one hot encoding for 'playlist_genre' column

I performed one hot encoding early in this stage since 'track_id' can have several unique 'playlist_genre' values. that way i wont loose data when dealing with dulicates (aggregating it by track_id)

```python

# Perform one-hot encoding for 'playlist_genre' while keeping NaN values
df1_dummies = pd.get_dummies(df[['track_id', 'playlist_genre']], columns=['playlist_genre'], dummy_na=True)

# Aggregate the one-hot encoded values by 'track_id' to ensure multiple genres are reflected
df1_dummies = df1_dummies.groupby('track_id', as_index=False).max()

# Ensure only one-hot encoded columns are converted to integers (0 and 1)
one_hot_columns = [col for col in df1_dummies.columns if col != 'track_id']
df1_dummies[one_hot_columns] = df1_dummies[one_hot_columns].astype(int)

# Merge the dummy columns back into df1, dropping the original 'playlist_genre' column
df1 = df1.drop(columns=['playlist_genre'], errors='ignore').merge(df1_dummies, on='track_id', how='left')

# Display the updated df1
print(df1.head())  # Print first few rows to verify

# Optional: Save to a CSV file for inspection
# df1.to_csv("df1_updated.csv", index=False)

```

## clean text
script that normalizes text by lowercasing, removes stop words, punctuation, and special characters, but keeps numbers and parentheses:

```python
import re
import string
from nltk.corpus import stopwords
import nltk
```

```python
import string

# Define a basic list of stopwords
stop_words = set([
    "i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours",
    "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers",
    "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves",
    "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are",
    "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does",
    "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until",
    "while", "of", "at", "by", "for", "with", "about", "against", "between", "into",
    "through", "during", "before", "after", "above", "below", "to", "from", "up", "down",
    "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here",
    "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more",
    "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so",
    "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"
])

# Function to clean text (keep numbers and parentheses but remove '-')
def clean_text(text):
    if pd.isna(text):  # Handle missing values
        return text
    text = text.lower()  # Convert to lowercase
    text = text.replace('-', ' ')  # Replace hyphens with spaces
    text = re.sub(r'[^a-zA-Z0-9()\s]', '', text)  # Remove special characters except numbers and parentheses
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    text = " ".join([word for word in text.split() if word not in stop_words])  # Remove stopwords
    return text

# Apply cleaning function to string columns, excluding 'track_id'
string_cols = [col for col in df1.select_dtypes(include=['object']).columns if col != 'track_id']
df1[string_cols] = df1[string_cols].apply(lambda col: col.apply(clean_text)).astype(str)

# Save the cleaned dataset
df1.to_csv("cleaned_spotify_songs.csv", index=False)

# Print the first few rows to confirm
print(df1.head())
```

# Handle  duplicates

```python
df1.duplicated(subset=['track_id']).sum()
```

```python

# Identify categorical and numeric columns
categorical_cols = df1.select_dtypes(include=['category', 'object']).columns.tolist()
numeric_cols = df1.select_dtypes(include=['number']).columns.tolist()

# Ensure 'track_album_release_date' is explicitly included in categorical columns
if 'track_album_release_date' not in categorical_cols:
    categorical_cols.append('track_album_release_date')

# Ensure 'track_id' is NOT treated as numeric
if 'track_id' in numeric_cols:
    numeric_cols.remove('track_id')

# 🔹 Ensure 'track_id' remains a string before grouping
df1['track_id'] = df1['track_id'].astype(str)

# Apply aggregation, treating 'track_id' as a categorical field
df1_grouped = df1.groupby('track_id', as_index=False).agg({
    **{col: 'max' for col in numeric_cols},  # Apply 'max' for numeric columns
    **{col: 'first' for col in categorical_cols},  # Keep first occurrence for categorical
    'track_id': 'first'  # Ensure 'track_id' remains unchanged by keeping the first value
})

# 🔹 Ensure 'track_id' remains in its original case
df1_grouped['track_id'] = df1_grouped['track_id'].astype(str)

# Re-sort the DataFrame to maintain order
df1_grouped = df1_grouped.sort_values(by='track_id').reset_index(drop=True)

# Display the updated DataFrame
print(df1_grouped.head())
```
